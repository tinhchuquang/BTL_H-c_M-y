```python
import csv
import numpy as np

def csv_dict_reader(file_name):
    with open(file_name) as in_file:
        X_train = np.zeros((50, 9))
        X_test = np.zeros((17, 9))
        Y_train = np.zeros((50,1))
        Y_test = np.zeros((17, 1))
        reader = csv.DictReader(in_file)
        i = 0 
        for row in reader:
            if i < 50:
                X_train[i:] = [1,row['lcavol'],row['lweight'],row['age'],row['lbph'],row['svi'],row['lcp'],row['gleason'],row['pgg45']]
                Y_train[i:] = [row['lpsa']]
            else:
                X_test[i-57:] = [1,row['lcavol'],row['lweight'],row['age'],row['lbph'],row['svi'],row['lcp'],row['gleason'],row['pgg45']]
                Y_test[i-50:] = [row['lpsa']]
            i+= 1
    return X_train, X_test, Y_train, Y_test

(X_train, X_test, Y_train, Y_test) = csv_dict_reader('1-prostate-training-data.csv')
```

```python
I = np.ones((9,9))
for i in range(9):
    for j in range(9):
        if i != j:
            I[i][j]=0
            
print(I)
```

```python
def caculator_w(lamda, A, y):
    B = A.T.dot(A) + I.dot(lamda)
    C = A.T.dot(y)
    w = np.dot(np.linalg.pinv(B), C)
    return w

def test(w, x_test, y_test):
    E = x_test.dot(w)
    F = E - y_test
    Error = 0
    for i in range(F.shape[0]):
        Error += F[i][0]*F[i][0]
    Error = np.sqrt(Error)
    return Error

w = caculator_w(1, X_train, Y_train)
print(w)
error = test(w, X_test, Y_test)
print(error)
```

```python
lamda = 0.001
w = caculator_w(lamda, X_train, Y_train)
error2 = error1 = test(w,X_test,Y_test)
    
for i in range(1000):
    lamda += 0.01
    w = w = caculator_w(lamda, X_train, Y_train)
    error2 = test(w,X_test,Y_test)
    print(error, error2-error1)
    if error2-error1 < 0:
        print(lamda)
    error1 = error2
```
